```python
# Part 2 of exercise 4 
import matplotlib.pyplot as plt


temp_list2 = [79, 68, 83, 69, 85, 66, 88, 67, 86, 63, 84, 59, 80, 57]



plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.plot(temp_list2, c="red")
ax.set_title("Temperature Based on Day", fontsize = 24)
ax.set_xlabel("Days of the week", fontsize = 18)
ax.set_ylabel('Temps', fontsize = 18)
plt.xticks([0,2,4,6,8,10,12], ['Wed', 'Tu', 'Fri', 'Sat', 'Sun', 'Mon', 'Tu'])
ax.tick_params(axis = 'both', which = 'major', labelsize = 16)



plt.show()
#  referenced to the week05 code for this problem 
```
