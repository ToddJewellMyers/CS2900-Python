```python
import csv
import matplotlib.pyplot as plt
import numpy as np


with open('crash_catalonia.csv') as File:
    reader = csv.reader(File)
    header = next(reader)
    crashes = []
    weeks = []
    for row in reader:
        crash = int(row[1])
        crashes.append(crash)
        week = row[0]
        weeks.append(week)

plt.style.use('ggplot')
bins = np.arange(min(crashes), max(crashes) + 1, 1)
fig, ax = plt.subplots(1, 1, figsize=(10, 7), tight_layout=True)
#ax.hist(crashes, bins=82, density=True, align='right', orientation='horizontal', log=True)
ax.hist(crashes, bins=82)
plt.bar(crashes, height=[7])
#ax.plot(weeks, crashes, c="purple")
ax.set_title("Crashes through Week ", fontsize=24)
ax.set_xlabel("Number of Crashes", fontsize=18)
ax.set_ylabel('Days of The Week', fontsize=18)
plt.yticks([ 1, 2, 3, 4, 5, 6,7], ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'])
ax.tick_params(axis='both', which='major', labelsize=16)
#fig.autofmt_xdate()


plt.show()
```
